```python
# Importing libraries needed for forcasting

import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Create dataframe
data = pd.DataFrame({
    'ds': pd.date_range(start='2022-01-01', periods=365, freq='D'), # freq='D': The frequency of the dates, 'D' stands for daily.
    'y': list(range(365))  # Keep 'y' consistent with the length of 'ds'
})

print('This is the head of the data', data.head())
print('This is the tail of the data', data.tail())

# Set the 'ds' column as the index
data.set_index('ds', inplace=True)

# Create an ExponentialSmoothing model instance with additive trend and no seasonality
model = sm.tsa.ExponentialSmoothing(data, trend='add', seasonal=None)

# Fit the model to the data
results = model.fit()

# Generate forecasts for the next 30 days
forecast = results.forecast(steps=30)
print('This is the forecast', forecast)

# Plot the original data and the forecast (optional)
plt.plot(data, label='Actual')
plt

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd

# Read the uploaded CSV file
data = pd.read_csv('Sales.csv')


# Print the first 10 rows to verify data loading
print(data.to_string)

# Define predictors and dependent variable
predictors = ['Profit', 'Sales', 'Discount']
dependent = 'Customer_Segment'

# Split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(data[predictors], data[dependent], test_size=0.2, random_state=42)

# Create a Logistic Regression model
model = LogisticRegression()

# Train the model on the training data
model.fit(train_data, train_labels)

# Make predictions on the test data
predictions = model.predict(test_data)

# Evaluate the accuracy of the model
accuracy = accuracy_score(test_labels, predictions)

# Print the accuracy of the trained model
print('Accuracy of the trained model: ', accuracy)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd

# Read the uploaded CSV file
data = pd.read_csv('Sales.csv')


# Print the first 10 rows to verify data loading
print(data.to_string)

# Define predictors and dependent variable
predictors = ['Profit', 'Sales', 'Order_Quantity']
dependent = 'Product_Category'

# Split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(data[predictors], data[dependent], test_size=0.2, random_state=42)

# Create a Logistic Regression model
model = LogisticRegression()

# Train the model on the training data
model.fit(train_data, train_labels)

# Make predictions on the test data
predictions = model.predict(test_data)

# Evaluate the accuracy of the model
accuracy = accuracy_score(test_labels, predictions)

# Print the accuracy of the trained model
print('Accuracy of the trained model: ', accuracy)
```
