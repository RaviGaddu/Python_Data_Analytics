```python

from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

Linear Regression

```python
x = df[['Unit_Price']]
y = df[['Sales']]
```

```python
# Create a LinearRegression model instance
model = LinearRegression()

# Train (fit) the model with input features (x) and corresponding target values (y)
model.fit(x,y)

# Create another LinearRegression model instance (unnecessary if the first model is already created)
linear_regressor = LinearRegression()

# Train (fit) the second model with the same input features (x) and target values (y)
linear_regressor.fit(x,y)

# Use the trained model to make predictions on the input features (x) The predicted values are stored in the variable Y_pred
Y_pred = linear_regressor.predict(x)
```

```python
# Create a scatter plot with the input features (x) on the x-axis and target values (y) on the y-axis
plt.scatter(x,y)

# Plot a line graph based on the predicted values (Y_pred) from the linear regression model
# The line is shown in red for better visibility
plt.plot( x, Y_pred, color='red')
# Display the scatter plot and the line graph
plt.show()
```
